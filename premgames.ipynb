```python
matches["date"] = pd.to_datetime(matches["date"])
matches.dtypes
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes #converts venue from string to category to numeric
matches.head()
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes #converts opponent to a code
matches.head()
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex = True).astype("int") #uses string replacement to remove the colon and minute from the time and only be left with hour as an int
#(in the str.replace function, the :.+ means remove any characters after the colon)
matches.head()
```

```python
matches["day_code"] = matches["date"].dt.dayofweek #converts date to day of week numerically
matches.head()
```

```python
matches["target"] = matches['result'].astype("category").cat.codes #converts loss to 1, draw to 0, and win to 2
matches.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 150, min_samples_split = 10, random_state = 42, class_weight = 'balanced')
```

```python
train = matches[matches["date"] < "2022-01-01"]
test = matches[matches["date"] >= "2022-01-01"]
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed = "left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset= new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel(0)
matches_rolling.head()
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < "2022-01-01"]
    test = data[data["date"] >= "2022-01-01"]
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted = preds), index = test.index)
    precision = precision_score(test["target"], preds, average="weighted")
    return combined, precision, preds
```

```python
combined, precision, preds = make_predictions(matches_rolling, predictors + new_cols)
precision
combined
```

```python
odds_df = pd.read_csv("E0 (2).csv")
odds_df = odds_df[["Date", "HomeTeam", "AwayTeam", "PSCH", "PSCD", "PSCA"]]
odds_df.columns = ["date", "home_team", "away_team", "home_odds", "draw_odds", "away_odds"]
odds_df["date"] = pd.to_datetime(odds_df["date"], format="%d/%m/%Y")
odds_df.head()
```

```python
matches_rolling["team_std"] = matches_rolling["team"].map(mapping)
matches_rolling["opponent_std"] = matches_rolling["opponent"].map(mapping)
matches_rolling.head()
```

```python
odds_21_22 = pd.read_csv("E0 (1).csv")
odds_21_22 = odds_21_22[["Date", "HomeTeam", "AwayTeam", "PSCH", "PSCD", "PSCA"]]
odds_21_22.columns = ["date", "home_team", "away_team", "home_odds", "draw_odds", "away_odds"]
odds_21_22["date"] = pd.to_datetime(odds_21_22["date"], format="%d/%m/%Y")
odds_21_22.head()
```

```python
odds_df["home_team"] = odds_df["home_team"].map(mapping)
odds_df["away_team"] = odds_df["away_team"].map(mapping)

odds_21_22["home_team"] = odds_21_22["home_team"].map(mapping)
odds_21_22["away_team"] = odds_21_22["away_team"].map(mapping)

odds_21_22
```

```python
all_odds_df = pd.concat([odds_df, odds_21_22], ignore_index=True)

# Merge the matches_rolling dataframe with the combined odds dataframe
merged_df = matches_rolling.merge(all_odds_df,
                                  left_on=["date", "team_std", "opponent_std"],
                                  right_on=["date", "home_team", "away_team"],
                                  how="left")

# Now repeat the merge for away matches to capture odds where the team_std was the away_team
merged_df = merged_df.merge(all_odds_df,
                              left_on=["date", "team_std", "opponent_std"],
                              right_on=["date", "away_team", "home_team"],
                              how="left",
                              suffixes=('', '_away_odds'))

# Combine the odds columns where they exist (handling NaN values from left joins)
merged_df['home_odds'] = merged_df['home_odds'].fillna(merged_df['home_odds_away_odds'])
merged_df['draw_odds'] = merged_df['draw_odds'].fillna(merged_df['draw_odds_away_odds'])
merged_df['away_odds'] = merged_df['away_odds'].fillna(merged_df['away_odds_away_odds'])

# Drop the duplicate odds columns and the extra team columns from the second merge
merged_df = merged_df.drop(columns=['home_odds_away_odds', 'draw_odds_away_odds', 'away_odds_away_odds', 'home_team_away_odds', 'away_team_away_odds'], errors='ignore')

merged_df.describe()
```
